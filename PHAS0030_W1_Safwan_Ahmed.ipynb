Charges can exhibit electrostatic potentials and fields around them. These can be described by the two formulas: $ V = \frac{k \cdot q}{r} $ and $ \mathbf{E} = \frac{k \cdot q}{r^2} \hat{\mathbf{r}} $. Potential describes the the amount of energy needed per unit of electric charge to move a charge from infinity to a point in space. Electric fields decribe the force a positive unit charge feels in a point in space. For simplicity, we will set k = 1 as it still allows us to show the shape of the fields.  

For a monopole charge, the potential and field is circulary symmetrical around the charge, both of which tends towards infinity as you go closer to the charge and 0 as you go away from the charge. For negative charges, the sign is flipped (negative infinity as you reach the charge).  
When you you have mutiple like charges, the values for potential and vectors for field for each individual charge add up in each point of space, and a new potential space or vector space forms. When you have opposite charges, the potentials or vectors subtract instead, creating a new potential or vectors
space.

We will explore some arrangements here in this notebook. 

```python
# Import libraries

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def potential(charge,x0,y0,x,y):
    """
    This function takes in the position and charge (of a charge), as well as a coordinate grid and returns the coordinate grid 
    with the potential across the grid.
    
    Inputs:
    charge  Value of charge
    x0, y0  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated
    """
    r2D = np.sqrt((x-x0) **2 + (y-y0) ** 2) # calculates the magnitude of distance from a point to the charge
    
    potential = charge / r2D # Calculates the potential of each point on the field using the equation above (and k=1)
    
    return potential
```

### Graphing Electrostatic Potential 

Here we will explore what the potentials will look like for a monopole, dipole and 2 dipoles. The equation for electric potential is $ V = \frac{k \cdot q}{r} $.

```python
# Define variables for charge(s) and positions
q1 = 1
x1 = 10
y1 = 10
# Create grid of x and y positions for plotting
x = np.arange(20)
y = np.arange(20)
x2D, y2D = np.meshgrid(x,y)

# Call function for a single +1 charge and plot the graph
pot1 = potential(q1,x1,y1,x2D,y2D) 
plt.imshow(pot1)
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Potential(V) for a single +1 charge")
plt.colorbar(); # Add a colourbar
```

```python
# set values for a dipole

q1, q2 = 1, -1
x1, x2 = 5, 15
y1, y2 = 10, 10

pot1 = potential(q1,x1,y1,x2D,y2D) + potential(q2,x2,y2,x2D,y2D) # Ad  potentials together 
plt.imshow(pot1, origin = 'lower')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Potential(V) for a dipole")
plt.colorbar();
```

```python
# Set values for 2 dipoles

q1, q2, q3, q4 = 1, -1, -1, 1
x1, x2, x3, x4 = 5, 5, 15, 15
y1, y2, y3, y4 = 5, 15, 5, 15

pot1 = potential(q1,x1,y1,x2D,y2D) + potential(q2,x2,y2,x2D,y2D) + potential(q3,x3,y3,x2D,y2D) + potential(q4,x4,y4,x2D,y2D) 
plt.imshow(pot1, origin = 'lower')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Potential(V) for 4 charges, with like charges in opposite diagnols")
plt.colorbar();
```

### Graphing Electrostatic Field
We will graph for the same scenarios as above using the equation $ \mathbf{E} = \frac{k \cdot q}{r^2} \hat{\mathbf{r}} $. Since the field is a vector space, we will represent it using a vector graph. 

```python
def field(charge,x0,y0,x,y):
    """This function takes in the position and charge (of a charge), as well as a coordinate grid and returns a vector field
    with its x and y components as seperate 2D arrays. 
    
    Inputs:
    charge  Value of charge
    x0, y0  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated
    """    
    x_comp, y_comp = x - x0, y - y0
    magnitude = np.sqrt((x-x0) **2 + (y-y0) ** 2) ** 3 # This is a different form of the equation but it is 
                                                               # effectively the same as mentioned above
    x_potential, y_potential = charge * x_comp / magnitude, charge * y_comp / magnitude # Calculates the electric field vectors
                                                                                        # and returns their direction as x and 
    return x_potential, y_potential                                                     # y coordinates
```

```python
q1 = 1
x1, y1 = 10,10
# Calculate x and y components of field for one charge
Ex1, Ey1 = field(q1,x1,y1,x2D,y2D)
# Plot with quiver
plt.quiver(x2D, y2D, Ex1, Ey1);
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Electric Field for a single +1 charge");
```

```python
q1, q2 = 1, -1
x1, x2 = 5, 15
y1, y2 = 10, 10

# Plot for a dipole
Ex1, Ey1 = np.add(field(q1,x1,y1,x2D,y2D), field(q2, x2, y2, x2D, y2D)) # Add fields together
plt.quiver(x2D, y2D, Ex1, Ey1)
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Electric Field for a dipole");
```

```python
q1, q2, q3, q4 = 1, -1, -1, 1
x1, x2, x3, x4 = 5, 15, 5, 15
y1, y2, y3, y4 = 15, 15, 5, 5

# Plot for 2 dipoles
Ex1, Ey1 = np.add(field(q1,x1,y1,x2D,y2D),field(q2, x2, y2, x2D, y2D)) + np.add(field(q3,x3,y3,x2D,y2D),field(q4,x4,y4,x2D,y2D))
plt.quiver(x2D, y2D, Ex1, Ey1)
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Electric Field for two dipoles");
```

### Conclusion

The graphs above show potential and field decreases as you go further from a positive charge, with the sign flipped for a negative charge.
When two or more are present, the fields or potentials add up, giving us the magnitude and/or direction at each point, which allows us to visualise the potential or electric in a space due to to the charges.
